```python
# Dependencies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Read columns CSV
columns = "Resources/columns.csv"
columns_df = pd.read_csv(columns, encoding="ISO-8859-1")
columns_df
```

```python
#Read columns CSV
responses = "Resources/responses.csv"
responses_df = pd.read_csv(responses, encoding="ISO-8859-1")
responses_df
```

```python
freetime = "Resources/freetime.csv"
freetime_df = pd.read_csv(freetime, encoding="ISO-8859-1")
freetime_df
```

```python
responses_df["Entertainment spending"].head()
```

```python
responses_df[["Entertainment spending", "Spending on healthy eating", "shopping centres", "age", "Countryside, outdoors"]].head()
```

```python
grouped_df = freetime_df.groupby(["Age"])
main_df=grouped_df.mean()
main_df
```

```python
cdp=main_df[['Countryside, outdoors','Dancing','Passive sport']].copy()
ax = cdp.plot(style=['o','rx','bx'],linestyle='solid',title="Activity Ratings by Age")
ax.set_xlabel("Age")
ax.set_ylabel("Rating Preferred")
```

```python
cdp=main_df[['Healthy eating','Shopping centres','Entertainment spending']].copy()
ax = cdp.plot(style=['rx','bo','go'],linestyle='solid', title="Activity Ratings by Age")
ax.set_xlabel("Age")
ax.set_ylabel("Rating Preferred")
```
